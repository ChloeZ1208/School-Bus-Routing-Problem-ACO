```python
# get the osmid of each point based on their coordinates
import osmnx as ox
import pandas as pd
north_toronto_chris_school = (43.7734184458907, -79.3368615141682)
G = ox.graph_from_point(north_toronto_chris_school, dist=3500, network_type='drive', clean_periphery=True, simplify=True)
fig, ax = ox.plot_graph(G)
```

```python
df = pd.read_excel (r'North toronto christian school_route1.xlsx')
coordinates = df.iloc[0:,1:3].values
```

```python
for x, y in coordinates:
    node = (x, y)
    print(ox.distance.get_nearest_node(G, node))
```
